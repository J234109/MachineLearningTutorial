```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PiDeMontecarlo(n, numeroDeExperimentos): 
    x = range(0,n)
    data = []
    for j in range(numeroDeExperimentos):
        value=0.0
        data1 = np.random.uniform(0,1,n)
        data2 = np.random.uniform(0,1,n)
        for i in x:
            if (data1[i] * data1[i] + data2[i] * data2[i])<1 :
                value+=1
        a = 4 * float(value) / n
        data.append(a) 
        df = pd.DataFrame(data)
        print (df.median())
        
    return data
```

```python
PiDeMontecarlo(100, 2)
%matplotlib inline
plt.hist(data)
```
